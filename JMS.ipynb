```python
def showbook(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html,'html.parser') 
    try:
        pages=int(soup.select('.number-of-pages')[0].text)  # 該分類共有多少頁
        print("共有",pages,"頁")
        for page in range(1,pages+1):
            pageurl=url[:len(url)-1] + '/' + str(page).strip()
            print("第",page,"頁",pageurl)
            showpage(pageurl)
    except:  # 沒有分頁的處理
        showpage(url)        

def showpage(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html,'html.parser') 
    #近期新書、在 class="mod type02_m012clearfix" 中
    #res = soup.find_all('div',{'class':'toc'})[0]
    head = soup.find('h1',{'id':'title'}).text
    res = soup.find_all('div',{'class':'toc-item no-access'})
    #print(res)  

    n=0  # 計算該分頁共有多少篇papers
    for item in res:
        con_type=item('p',{'class':'content-type'})[0].text
        title=item('h3',{'class':'title'})[0].text
        try:
            con_url= 'https://link.springer.com' + str(item.select('a')[1].get('href')).strip()
        except:  # 沒有分頁的處理
            con_url= 'https://link.springer.com' + str(item.select('a')[0].get('href')).strip()
        
        down_url = 'http://sci-hub.tw/' + con_url
        authors=item('span',{'class':'authors'})[0].text
       # article=item('small',{'class':'citation-info'})[0].text
       # print(con_type)
        #print(title)
        #print(re.sub('\s','',authors))
        #print(article)
        #print(con_url)  
        n+=1
        #print("n=",n)
        # 將資料加入 list1 串列中
        listdata=[head,con_type,title,authors,con_url,down_url]
        list1.append(listdata)        

def twobyte(kindno):
    if kindno<10:
        kindnostr="0"+str(kindno)
    else:
        kindnostr=str(kindno) 
    return kindnostr

#主程式
import requests
from bs4 import BeautifulSoup 
import re
import openpyxl        
workbook=openpyxl.Workbook()   #建立一個工作簿
sheet = workbook.worksheets[0] #獲取工作表 
list1=[] 

kindno=1  # 計算共有多少分類  
volume=42
issue = 1
homeurl = 'https://link.springer.com/journal/volumesAndIssues/10916'

html = requests.get(homeurl).text
#print(html)
soup = BeautifulSoup(html,'html.parser') 

#res = soup.find('div',{'id':'volumes-list'})
#hrefs=res.select("h2")
hrefs = soup.find_all('li',{'class':'issue-item'})
for href in hrefs: 
    issue_url= 'https://link.springer.com' + str(href.select('a')[0].get('href')).strip()
    print('issue_url=',issue_url)
    #kindurl = url
 #   print("\nkindno=",kindno)  
#    kind=href.text #分類
#    showpage(issue_url) # 顯示該分類所有書籍
    showbook(issue_url)
#    if kindno==2: break  # 開發階段    

# excel 資料
listtitle=["Header","Topical","title","authors","url","down_url"]
sheet.append(listtitle)  # 標題
for item1 in list1: #資料
    sheet.append(item1) 
    
workbook.save('JMS_all.xlsx')    
```
