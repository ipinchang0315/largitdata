```
import requests
from bs4 import BeautifulSoup
import openpyxl
workbook=openpyxl.Workbook()
sheet = workbook.worksheets[0]
list1=[]

for i in range(27):
  homeurl= 'https://tjn.moe.edu.tw/jobshow.php?pg=' + str(i)
  html = requests.get(homeurl)
  html.encoding = 'utf-8'  
  html = html.text
  #print(html)
  soup = BeautifulSoup(html,'html.parser')
  soup.findAll('table')[0].tbody.findAll('tr')
  for row in soup.findAll('table')[0].tbody.findAll('tr'):
    first_column = row.findAll('td')[0].text
    two_column = row.findAll('td')[1].text
    third_column = row.findAll('td')[2].text
    four_column = row.findAll('td')[3].text
    five_column = row.findAll('td')[4].text
    six_column = row.findAll('td')[5].text
    url= 'https://tjn.moe.edu.tw/' + row.findAll('a')[0]["href"]
    listdata=[first_column, two_column, third_column, five_column, six_column, url]
    list1.append(listdata)
    sheet.append(listdata)
    print(first_column, two_column, third_column, five_column, six_column, url)

workbook.save('job.xlsx')
```
